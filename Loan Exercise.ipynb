```python
# (b) create a repayment lookup: 
import pandas as pd
from io import StringIO
from datetime import date

def expand_dates(ser):
    return pd.DataFrame({'SupposedRepaymentMonth': pd.date_range(ser['Loan Start Date'].min(), '2021-03-31', freq='M')})

loanextended = loan.groupby(['Loan_ID']).apply(expand_dates).reset_index()\
          .merge(loan, how='left')[['Loan_ID','Loan Start Date','Loan Start Month', 'SupposedRepaymentMonth', 'Loan Amount']].ffill()
loanextended['SupposedRepaymentMonth'] = loanextended['SupposedRepaymentMonth'].dt.strftime('%Y-%m')
```

```python
# (b)
mth_repayment = repayment.groupby(['Repayment Month','Loan ID'])['Repayment Amount'].sum().reset_index()
mth_df = pd.merge(loanextended, mth_repayment,  how='left', left_on=['SupposedRepaymentMonth','Loan_ID'], right_on = ['Repayment Month','Loan ID']).sort_values(by=['Loan_ID','SupposedRepaymentMonth'])
mth_df['Repayment Amount'] = mth_df['Repayment Amount'].fillna(0)
mth_df['RepaymentAccumulation'] = mth_df.groupby(['Loan_ID'])['Repayment Amount'].cumsum(axis=0).ffill()
mth_df['outstanding'] = mth_df['Loan Amount'] - mth_df['RepaymentAccumulation'].fillna(0)
mth_df['%outstanding'] = mth_df['outstanding'] / mth_df['Loan Amount'] *100
mth_df.drop(['Loan ID'], axis = 1, inplace = True)
#test = mth_df.copy()
mth_df
```
